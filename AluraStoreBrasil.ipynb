### Importação dos dados



```python
# Importar bibliotecas necessárias:
import pandas as pd
import matplotlib.pyplot as plt

url1= "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja1 = pd.read_csv(url1)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

# Adicionar uma coluna de identificação
loja1["Loja"] = "Loja 1"
loja2["Loja"] = "Loja 2"
loja3["Loja"] = "Loja 3"
loja4["Loja"] = "Loja 4"

# Unificar todas em um único DataFrame
lojas = pd.concat([loja1, loja2, loja3, loja4])

# Visualizar as primeiras linhas
lojas.head()
```

```python
#Observar a quantidade de linhas e colunas
lojas.shape
```

```python
#Explorar os nomes das colunas
lojas.columns
```

```python
#Informações e tipo de cada coluna
lojas.info()
```

```python
#Exibir categorias e produtos sem repetição, em ordem alfabética
lojas[['Categoria do Produto', 'Produto']].drop_duplicates().sort_values(['Categoria do Produto', 'Produto'])


```

```python
#Contar quantos produtos únicos há em cada categoria
lojas.groupby('Categoria do Produto')['Produto'].nunique().reset_index()
```

#1. Análise do faturamento



```python
# Faturamento total por loja (somando os preços)
faturamento_por_loja = lojas.groupby('Loja')['Preço'].sum().reset_index()

```

```python
# Renomear as colunas para ficar mais claro
faturamento_por_loja.columns = ['Loja', 'Faturamento Total (R$)']

```

```python
# Exibir resultado formatado
faturamento_por_loja
```

# 2. Vendas por Categoria

```python
# Agrupar por loja e categoria e contar o número de vendas (linhas)
vendas_por_categoria = lojas.groupby(['Loja', 'Categoria do Produto']).size().reset_index(name='Quantidade Vendida')
```

```python
# Ordenar da maior para a menor quantidade
vendas_por_categoria = vendas_por_categoria.sort_values(['Loja', 'Quantidade Vendida'], ascending=[True, False])


```

```python
# Exibir resultado
vendas_por_categoria
```

```python
# Selecionar as 3 categorias mais vendidas de cada loja
top3_categorias_por_loja = vendas_por_categoria.groupby('Loja').head(3).reset_index(drop=True)

# Exibir resultado
top3_categorias_por_loja
```

# 3. Média de Avaliação das Lojas


```python
# Calcular a média das avaliações por loja (já arredondando para 2 casas decimais)
avaliacao_media_por_loja = lojas.groupby('Loja')['Avaliação da compra'].mean().round(2).reset_index()

# Renomear as colunas para clareza
avaliacao_media_por_loja.columns = ['Loja', 'Média das Avaliações']


```

```python
# Exibir resultado
avaliacao_media_por_loja
```

# 4. Produtos Mais e Menos Vendidos

```python
# Agrupar por loja e produto, contando o número de vendas
vendas_por_produto = lojas.groupby(['Loja', 'Produto']).size().reset_index(name='Quantidade Vendida')

# Identificar o produto mais vendido de cada loja
mais_vendidos = vendas_por_produto.loc[vendas_por_produto.groupby('Loja')['Quantidade Vendida'].idxmax()].reset_index(drop=True)
mais_vendidos = mais_vendidos.rename(columns={
    'Produto': 'Produto Mais Vendido',
    'Quantidade Vendida': 'Qtd Mais Vendido'
})

# Identificar o produto menos vendido de cada loja
menos_vendidos = vendas_por_produto.loc[vendas_por_produto.groupby('Loja')['Quantidade Vendida'].idxmin()].reset_index(drop=True)
menos_vendidos = menos_vendidos.rename(columns={
    'Produto': 'Produto Menos Vendido',
    'Quantidade Vendida': 'Qtd Menos Vendido'
})

```

```python
# Juntar (merge) as duas tabelas lado a lado
resumo_vendas = pd.merge(mais_vendidos, menos_vendidos, on='Loja')

# Exibir resultado
resumo_vendas
```

# 5. Frete Médio por Loja

```python
# Calcular o custo médio de frete por loja (com duas casas decimais)
frete_medio_por_loja = lojas.groupby('Loja')['Frete'].mean().round(2).reset_index()

# Renomear as colunas para clareza
frete_medio_por_loja.columns = ['Loja', 'Custo Médio de Frete (R$)']

```

```python
# Exibir resultado
frete_medio_por_loja

```

<font size="5">6. Gráfico do Faturamento.</font>

```python
# Gráfico direto do DataFrame faturamento_por_loja
plt.figure(figsize=(10, 6))
plt.barh(faturamento_por_loja['Loja'], faturamento_por_loja['Faturamento Total (R$)'], color='teal')

# Título e rótulos
plt.title('Faturamento Total por Loja', fontsize=14, fontweight='bold')
plt.xlabel('Faturamento (R$)', fontsize=12)
plt.ylabel('Lojas', fontsize=12)

# Adicionar valores ao lado das barras
for i, valor in enumerate(faturamento_por_loja['Faturamento Total (R$)']):
    plt.text(valor + 5000, i, f"R$ {valor:,.2f}".replace(",", "X").replace(".", ",").replace("X", "."), va='center')

plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

<font size="5">7. Gráfico Avaliação do Cliente.</font>

```python
# Gráfico de pizza das avaliações médias com valores
plt.figure(figsize=(8, 8))

# Criação do gráfico
plt.pie(
    avaliacao_media_por_loja['Média das Avaliações'],
    labels=[f"{loja} - {nota:.2f}" for loja, nota in zip(avaliacao_media_por_loja['Loja'], avaliacao_media_por_loja['Média das Avaliações'])],
    startangle=90,
    colors=['#1E88E5', '#FB8C00', '#43A047', '#E53935'],
    wedgeprops={'edgecolor': 'white'},
    autopct=lambda p: f"{p:.1f}%" if p > 0 else ''  # Mostra % apenas como referência visual
)

# Título
plt.title('Distribuição Visual das Avaliações Médias por Loja', fontsize=14, fontweight='bold')

# Exibição
plt.show()
```

<font size="5">8. Gráfico Custo do Frete.</font>

```python
# Gráfico de área para o frete
plt.figure(figsize=(9, 6))
plt.fill_between(
    frete_medio_por_loja['Loja'],
    frete_medio_por_loja['Custo Médio de Frete (R$)'],
    color='#00ACC1', alpha=0.4
)
plt.plot(
    frete_medio_por_loja['Loja'],
    frete_medio_por_loja['Custo Médio de Frete (R$)'],
    color='#006064', marker='o', linewidth=2
)

# Título e rótulos
plt.title('Custo Médio de Frete por Loja', fontsize=14, fontweight='bold')
plt.xlabel('Lojas', fontsize=12)
plt.ylabel('Custo Médio de Frete (R$)', fontsize=12)

# Adicionar rótulos sobre os pontos
for i, valor in enumerate(frete_medio_por_loja['Custo Médio de Frete (R$)']):
    plt.text(i, valor + 0.1, f"R$ {valor:.2f}".replace('.', ','),
             ha='center', fontsize=10, fontweight='bold')

# Grade e exibição
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
